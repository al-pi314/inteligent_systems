```python
import numpy as np
from matplotlib import pyplot as plt
from random import choice, seed
seed(100)
```

```python
# maze_example.txt
moves = {"L": 1, "R": 2, "D": 3, "U": 4}
# correct_path = [moves["L"], moves["U"], moves["L"], moves["L"],  moves["U"],  moves["U"],  moves["U"]]
correct_path = [choice(list(moves.values())) for i in range(900)]
indexes = [(i+1)/1000 for i in range(len(correct_path))]
plt.scatter(indexes, correct_path)
```

```python
my_polynomial = np.polyfit(indexes, correct_path, 300)
my_polynomial
```

```python
x = np.linspace(1, 7, 100)
plt.plot(x, [np.polyval(my_polynomial, i) for i in x])
poly_values = [np.polyval(my_polynomial, i) for i in indexes]
```

```python
# error
[correct_path[i] - poly_values[i] for i in range(len(indexes))]
```
